```python
from qiskit.visualization import plot_histogram

sim_counts  = {'00': 472, '11': 528}                          # Aer result
real_counts = {'00': 510, '11': 427, '10': 35, '01': 28}      # ibm_fez result

plot_histogram([sim_counts, real_counts],
               legend=['Simulator (ideal)', 'ibm_fez (real)'],
               title='Bell state: ideal vs. real hardware')
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()
qc.draw('mpl')
```

```python
errors = real_counts.get('01', 0) + real_counts.get('10', 0)
print(f"Error rate: {errors/1000:.1%}")
```
